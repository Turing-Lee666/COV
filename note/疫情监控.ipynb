#### urllilb的基本使用

````python
from urllib import request  # from 包 import 模块

url = "http://www.baidu.com"
res = request.urlopen(url)  # 调用request模块的urlopen()函数，可以接收一个参数，是一个url地址,返回一个响应对象，用res这个变量来接收他。即获取响应。

# 拿到响应之后就可以打印出几个有用的信息
print(res.info())  # 响应头
print(res.getcode())  # 状态码 2xx(网页正常情况)，3xx(发生了重定向，已访问了A，实际上A在内部又给你转发到了B网站)，4xx(表示访问的资源有问题，比如：404访问的这个页面不存在)，5xx(服务器内部接受到这个请求之后，执行发生了内部的错误，服务器崩了)
print(res.geturl())  # 从res对象里面拿到返回响应地址
````

````python
html = res.read()  # 用html这个变量来接收网页的源码，但是这时候返回的是字节码的形式
# print(html)
html = html.decode("utf-8")  # 把原始的html使用decode()做一个解码操作,指定我们的编码叫做utf-8
print(html)
````

````python
# 添加 header 信息，这是最基本的反爬措施
url = "http://www.dianping.com"
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36"
}
req = request.Request(url, headers = header)  #用request里面Request类封装一个req请求对象，接收的参数一个是url,另外一个是headers
res = request.urlopen(req)  # 调用request模块的urlopen()函数，可以接收一个参数，是一个url地址,返回一个响应对象，用res这个变量来接收他。即获取响应。

# 拿到响应之后就可以打印出几个有用的信息
print(res.info())  # 响应头
print(res.getcode())  # 状态码 2xx(网页正常情况)，3xx(发生了重定向，已访问了A，实际上A在内部又给你转发到了B网站)，4xx(表示访问的资源有问题，比如：404访问的这个页面不存在)，5xx(服务器内部接受到这个请求之后，执行发生了内部的错误，服务器崩了)
print(res.geturl())  # 从res对象里面拿到返回响应地址
````

#### 使用re解析

````python
# 使用正则表达式来做具体精确数字的匹配

import re
text = context.text  # 拿到context对象的text属性的值（p标签里的文本值）
print(text)
patten = "新增确诊病例(\d+)例"  #  首先传入一个正则表达式  \d匹配一个数字字符。等价于[0-9]。  +匹配前面的子表达式一次或多次(大于等于1次）。  这儿已经使用了分组。
res = re.search(patten, text)  # 调用re的search()函数,该函数的功能就是匹配  他里面要传参，第一个是正则表达式，第二个参数就是待匹配的字符串 
print(res)
````

````python
print(res.groups())  # 拿到所有 ()即分组 里面匹配到的值  返回值以一个元组的形式返回    
print(res.group(0))  # 0表示我们匹配到的字符串本身
print(res.group(1))  # 1就是第一个分组匹配的内容，2就是第二个分组匹配的内容，以此类推
````

````python
patten = "新增确诊病例(\d+).*?出院病例(\d+).*?疑似病例(\d+).*?死亡病例(\d+)"  # .*来表示匹配中间的任意个字符(0到多个)  ？将贪婪模式变成非贪婪模式，这样的话它就会匹配的越少越好
res = re.search(patten, text)
'''1. 在text中查找满足条件的字符串，匹配不上返回None
    2. 对返回结果(匹配结果)可以分组(分组的操作就是在正则表达式上面加上一些())，可在字符串内添加小括号()分离数据:
        groups():  可以拿到所有分组即()里面的内容
        group(index): 返回指定分组内容  注意：group(0)匹配的永远都是完整的字符串本身
'''
print(res.groups())  # 拿到所有 ()即分组 里面匹配到的值

# group(index): 返回指定分组内容
print(res.group(0))  # 0表示我们匹配到的字符串本身
print(res.group(1),res.group(2),res.group(3),res.group(4))  # 1就是第一个分组匹配的内容，2就是第二个分组匹配的内容，以此类推
````

````python
import pymysql
import time
import json  
import traceback  # 追踪异常
````

````python
def get_tencent_data():
    """
    :return: 返回历史数据和当日详细数据
    """
    # 指定目标地址  从腾讯后台里面的console里面看到这个地址
    url0 = 'https://view.inews.qq.com/g2/getOnsInfo?name=disease_other'  # 这个是历史数据地址
    url = 'https://view.inews.qq.com/g2/getOnsInfo?name=disease_h5'  # 这个是当日详细数据地址
    headers = {
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36'
    }  # 添加header   
    r0 = requests.get(url0,headers)  # 使用requests模块里的get()方法
    r = requests.get(url,headers)  # 使用requests模块里的get()方法
    
    res0 = json.loads(r0.text)  # 因为r0.text是一个json字符串，所以做第一级转换，把它转换成一个字典
    res = json.loads(r.text)  # 因为r.text是一个json字符串，所以做第一级转换，把它转换成一个字典
    
    data_all0 = json.loads(res0["data"])  # 将data键的数据拿到  json模块的loads()函数将json字符串转换为字典
    data_all = json.loads(res["data"])  # 将data键的数据拿到  json模块的loads()函数将json字符串转换为字典
    
    history = {}  # 历史数据
    for i in data_all0["chinaDayList"]:  # 遍历中国每日疫情数据列表  元素是中国每日疫情数据  元素类型是字典
        ds = "2020." + i["date"]
        tup = time.strptime(ds, "%Y.%m.%d")
        ds = time.strftime("%Y-%m-%d",tup)  # 改变时间格式，不然插入数据库会报错，数据库是datetime类型
        confirm = i["confirm"]
        suspect = i["suspect"]
        heal = i["heal"]
        dead = i["dead"]
        history[ds] = {"confirm":confirm, "suspect":suspect, "heal":heal, "dead":dead}  # 以每一天为这个字典的Key,然后Value是一个新的字典
    for i in data_all0["chinaDayAddList"]:  # 遍历每日新增列表
        ds = "2020." + i["date"]
        tup = time.strptime(ds, "%Y.%m.%d")
        ds = time.strftime("%Y-%m-%d",tup)  # 改变时间格式，不然插入数据库会报错，数据库是datetime类型
        confirm = i["confirm"]
        suspect = i["suspect"]
        heal = i["heal"]
        dead = i["dead"]
        history[ds].update({"confirm_add":confirm, "suspect_add":suspect, "heal_add":heal, "dead_add":dead})
        '''由于history字典的ds键的值依然是字典，因此我们对history[ds]字典进行更新，用update() 
        '''
    details = []  # 当日详细数据
    update_time = data_all["lastUpdateTime"]  # 上次数据更新时间
    data_country = data_all["areaTree"]  # list 25个国家
    data_province = data_country[0]["children"]  # data_coountry这个列表的0号元素是个字典代表中国  data_country[0]["children"]代表中国各省
    for pro_infos in data_province:  # 再遍历各省份的数据  data_province应该是个列表
        province = pro_infos["name"]  # 省名
        for city_infos in pro_infos["children"]:  # 省里面还有"children"键  就是具体的某一个市  如果要是直辖市的话就是具体某一个区
            city = city_infos["name"]
            confirm = city_infos["total"]["confirm"]
            confirm_add = city_infos["today"]["confirm"]
            heal = city_infos["total"]["heal"]
            dead = city_infos["total"]["dead"]
            details.append([update_time, province, city, confirm, confirm_add, heal, dead])
    return history, details
````

````python
def get_conn():
    """
    :return: 连接，游标
    """
    # 创建连接
    conn = pymysql.connect(host="127.0.0.1",
                           user="root",
                           password="",
                           db="cov",
                           charset="utf8")
    # 创建游标
    cursor = conn.cursor()  # 执行完毕返回的结果集默认以元组显示
    return conn, cursor

def close_conn(conn, cursor):
    if cursor:  # 如果cursor存在
        cursor.close()  # 关闭游标
    if conn:  # 如果conn存在
        conn.close()  # 关闭连接
        
        
````

````python
def update_details():
    """
    更新 details 表(插入到详情表)
    :return:
    """
    cursor = None
    conn = None
    try:
        li = get_tencent_data()[1]  
        '''利用get_tencent_data()这个函数的返回值里面的下标为1的那个  0 是历史数据字典history，1 最新详细数据列表details
        '''
        conn, cursor = get_conn()
        sql = "insert into details(update_time,province,city,confirm,confirm_add,heal,dead) values(%s,%s,%s,%s,%s,%s,%s)"
        """插入语句：insert into details(涉及到的字段名) values(具体的值，使用%s来做占位)
        """
        
        sql_query = 'select %s=(select update_time from details order by id desc limit 1)'  # 对比当前最大时间戳 
        """在sql语句里面“=”表示判断是否相等
        ()里面是一个子查询
        order by id desc 按id这个主键降序排序
        limit 1 只显示一条
        拿到最后一条记录里面的最近更新时间update_time  拿到这个时间和传入的时间(即%s所代表的时间)来做一个比较   
       
        """
        cursor.execute(sql_query,li[0][0])  # 
        """传到sql_query这条sql语句里%s对应的值是li[0][0]
        li表示 最新详细数据列表details
        li[0] details列表的0号元素  仍然是一个列表，长这样[update_time, province, city, confirm, confirm_add, heal, dead]  这里记为第一条记录
        li[0][0] details列表0号元素的0号元素  就是update_time,即上次数据更新时间
        这样执行下来之后，如果说details数据表中最后更新的时间和我们传入的时间是一致的那就说明这个数据还没有更新，就没必要往details数据表里面插
        如果说两个时间相等的话cursor.fetchone()[0]就返回true，不相等的话就会返回false
        """
        if not cursor.fetchone()[0]:  
            print(f"{time.asctime()}开始更新最新数据")
            for item in li:
                cursor.execute(sql,item)  # 执行插入语句来更新details表
            conn.commit()  # 提交事务 update(改) delete(删) insert(增)操作
            print(f"{time.asctime()}更新最新数据完毕")
        else:
            print(f"{time.asctime()}已是最新数据！")
    except:
        traceback.print_exc()  # 在except里面使用了一个追踪异常的方法  把我们发生异常的调用栈信息打印出来  哪一行出错了，什么原因引起的
    finally:
        close_conn(conn, cursor)  # 关闭连接（游标和连接）
````

````python
def insert_history():
    """
        插入历史数据表
    :return:
    """
    cursor = None
    conn = None
    try:
        dic = get_tencent_data()[0]  # 0 是历史数据字典,1 最新详细数据列表
        print(f"{time.asctime()}开始插入历史数据")
        conn, cursor = get_conn()
        sql = "insert into history values(%s,%s,%s,%s,%s,%s,%s,%s,%s)"
        for k, v in dic.items():
            # item 格式 {'2020-01-13': {'confirm': 41, 'suspect': 0, 'heal': 0, 'dead': 1}
            cursor.execute(sql, [k, v.get("confirm"), v.get("confirm_add"), v.get("suspect"),
                                 v.get("suspect_add"), v.get("heal"), v.get("heal_add"),
                                 v.get("dead"), v.get("dead_add")])

        conn.commit()  # 提交事务 update delete insert操作
        print(f"{time.asctime()}插入历史数据完毕")
    except:
        traceback.print_exc()
    finally:
        close_conn(conn, cursor)
````

````python
def update_history():
    """
    更新历史数据
    :return:
    """
    cursor = None
    conn = None
    try:
        dic = get_tencent_data()[0]  #  0 是历史数据字典,1 最新详细数据列表
        print(f"{time.asctime()}开始更新历史数据")
        conn, cursor = get_conn()
        sql = "insert into history values(%s,%s,%s,%s,%s,%s,%s,%s,%s)"
        sql_query = "select confirm from history where ds=%s"
        for k, v in dic.items():
            # item 格式 {'2020-01-13': {'confirm': 41, 'suspect': 0, 'heal': 0, 'dead': 1}
            if not cursor.execute(sql_query, k):
                cursor.execute(sql, [k, v.get("confirm"), v.get("confirm_add"), v.get("suspect"),
                                     v.get("suspect_add"), v.get("heal"), v.get("heal_add"),
                                     v.get("dead"), v.get("dead_add")])
        conn.commit()  # 提交事务 update delete insert操作
        print(f"{time.asctime()}历史数据更新完毕")
    except:
        traceback.print_exc()
    finally:
        close_conn(conn, cursor)
````

### 存储百度数据

````python
def get_baidu_hot():
    """
    :return: 返回百度疫情热搜
    """
    option = ChromeOptions()  
    option.add_argument("--headless")  # 隐藏浏览器
    option.add_argument('--no-sandbox')
    
    url = "https://voice.baidu.com/act/virussearch/virussearch?from=osari_map&tab=0&infomore=1"
    browser = Chrome(options=option,executable_path="./chromedriver.exe")
    browser.get(url)
    # 找到展开按钮（通过xpath路径）
    dl = browser.find_element_by_xpath('//*[@id="ptab-0"]/div/div[2]/section/div')
    dl.click()  # 点击该按钮，以展开剩余部分
    time.sleep(1)
    # 找到热搜标签
    c = browser.find_elements_by_xpath('//*[@id="ptab-0"]/div/div[2]/section/a/div/span[2]')  # 找到的20条热搜标签以列表的形式返回
    context = [i.text for i in c]  # 获取标签内容  这是一个列表生成器  列表元素是i.text即每一条热搜标签的文本内容  
    print(context)
    return context


def update_hotsearch():
    """
    将疫情热搜插入数据库
    :return:
    """
    cursor = None
    conn = None
    try:
        context = get_baidu_hot()  # 拿取到热搜内容，有20条数据
        print(f"{time.asctime()}开始更新热搜数据")  # 记录一下开始更新热搜数据的这个时间
        conn, cursor = get_conn() 
        sql = "insert into hotsearch(dt,content) values(%s,%s)"
        ts = time.strftime("%Y-%m-%d %X")  # 把时间处理成：年月日时分秒 这种格式
        for i in context:  # 遍历拿到的context列表
            cursor.execute(sql,(ts, i))  # 插入数据（时间，热搜内容）
        conn.commit()  # 提交事务保存数据  针对：增删改
        print(f"{time.asctime()}数据更新完毕")#  最后输出一下热搜数据更新结束的时间
    except:
        traceback.print_exc()
    finally:
        close_conn(conn, cursor)
````
